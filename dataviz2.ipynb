# Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset

```python
data=pd.read_csv('jobpredict.csv')
```

```python
data.info()
```

# Distribution plot

```python
sns.distplot(data['ssc_p'])
```

```python
sns.distplot(data['hsc_p'])
```

```python
sns.distplot(data['degree_p'])
```

```python
sns.distplot(data['salary'])
```

# Joint plot

```python
sns.jointplot(x='ssc_p',y='hsc_p',data=data,kind='scatter')
```

```python
sns.jointplot(x='hsc_p',y='degree_p',data=data,kind='hex')
```

# Pair plot

```python
sns.pairplot(data)
```

# Bar plot

```python
plt.figure(figsize=(18,6))
sns.barplot(x='Jobprofile',y='salary',data=data)
plt.tight_layout()
plt.savefig('bar1.png')
```

```python
sns.barplot(x='Majoredu',y='salary',data=data)
```

```python
sns.barplot(x='ssc_b',y='ssc_p',data=data)
```

```python
sns.barplot(x='hsc_b',y='hsc_p',data=data)
```

# Count plot

# Box plot

```python
sns.boxplot(x="ssc_p", y="Jobprofile", data=data,palette='rainbow')
```

```python
sns.boxplot(x="hsc_p", y="Jobprofile", data=data,palette='rainbow')
```

```python
sns.boxplot(x="degree_p", y="Jobprofile", data=data,palette='rainbow')
```

# Violin plot

```python
sns.violinplot(x="etest_p", y="Jobprofile", data=data,palette='rainbow')
plt.savefig('vilion1.png')
```

# Swarm plot

```python
plt.figure(figsize=(18,5))
sns.swarmplot(x="Jobprofile", y="salary",hue='gender',data=data, palette="Set1")
```

# Heat map

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr())
plt.savefig('heat1.png')
```
